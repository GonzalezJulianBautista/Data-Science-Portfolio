```python
# We import the libraries to use. // Importamos las variables pertinentes. 
import pandas as pd
import numpy as np
import matplotlib as plt
import chart_studio.plotly as py
import seaborn as sns
import math
from numpy.random import randn
np.random.seed(101)
import cufflinks as cf
%matplotlib inline
```

```python
# Using Pandas, we read and load the CSV containing the 'Degrees' Dataset. 
# The 'Degrees' variable is created, containing the table 'Degrees'.
# Cargamos el CSV con la tabla 'Degrees', usando Pandas. 
# Asignamos la tabla a la variable 'degrees'.
degrees = pd.read_csv("Degrees Dataset.csv", delimiter = ';')
degrees
```

```python
# We calculate the initial entropy of the set, given by the formula 
# -Xlog2X for each class of the set.

# Calculamos la entropía inicial de la tabla 'Degrees', 
# aplicando la fórmula -Xlog2X a cada clase del set. 

# X = instances of a given class / total instances in the whole set. 
# X = sucesos de una determinada clase / sucesos totales del set. 

# Entropía: Medida de la incertidumbre existente ante un conjunto de datos.
# Entropy: lack of order or predictability; gradual decline into disorder.

# This initial entropy will be asigned to the variable 'SoftEng_Estart'.
# We will return to this variable at the very end of this analysis. 

# La entropía inicial obtenida va a ser asginada a la variable 'Degrees_Estart'.
# Vamos a volver sobre esta variable en el segmento final de este análisis. 

number_instances_SoftEng_A = sum(degrees['SoftEng'] == 'A') 
number_instances_SoftEng_B = sum(degrees['SoftEng'] == 'B')
number_total_instances = number_instances_SoftEng_A + number_instances_SoftEng_B
number_instances_class_FIRST = sum(degrees['Class'] == 'FIRST')
number_instances_class_SECOND = sum(degrees['Class'] == 'SECOND')
entropy_instances_class_FIRST = -(number_instances_class_FIRST/number_total_instances)*math.log2(number_instances_class_FIRST/number_total_instances)
entropy_instances_class_SECOND = -(number_instances_class_SECOND/number_total_instances)*math.log2(number_instances_class_SECOND/number_total_instances)
Degrees_Estart = entropy_instances_class_FIRST + entropy_instances_class_SECOND

print('For the degrees dataset, the initial entropy is: ',Degrees_Estart,)
```

```python
# First, we will calculate the InfoGain we earn when the table 'Degrees'
# is splitted by SoftEng.

# Para comenzar, vamos a calcular la InfoGain que conseguimos 
# cuando el split de la tabla Degrees es hecho en base a la columna SoftEng.
```

```python
# We print the information for SoftEng taking the value 'A'.
# This includes the total instances, number of classes and it's names. 

# Imprimimos en pantalla la información para SoftEng tomando por valor 'A'.
# Van a ser de utilidad la cantidad de sucesos, cantidad de clases y sus nombres. 
print( 'FOR SPLITTING IN SoftEng = A: ')   
print('Number of Instances: ', number_instances_SoftEng_A,)
print('Number of Classes: 2')
print('Classes: FIRST, SECOND')
```

```python
# We add up and print the instances for which SoftEng = A & CLASS = FIRST.

# Sumamos e imprimimos en pantalla la cantidad de instancias donde 
# SoftEng = A & CLASS = FIRST.

SoftEng_A_FIRST = pd.DataFrame(degrees[(degrees["SoftEng"] == 'A') & (degrees["Class"] == 'FIRST')])
SoftEng_A_FIRST = SoftEng_A_FIRST.reset_index(drop = True)
SoftEng_A_FIRST

index = SoftEng_A_FIRST.index
number_of_rows_SoftEng_A_FIRST = len(index)
print('SoftEng = A, Class = FIRST, Number Of Rows = ',number_of_rows_SoftEng_A_FIRST,)
```

```python
# We add up and print the instances for which SoftEng = A & CLASS = SECOND.

# Sumamos e imprimimos en pantalla la cantidad de instancias donde 
# SoftEng = A & CLASS = SECOND.

SoftEng_A_SECOND = pd.DataFrame(degrees[(degrees["SoftEng"] == 'A') & (degrees["Class"] == 'SECOND')])
SoftEng_A_SECOND = SoftEng_A_SECOND.reset_index(drop = True)
SoftEng_A_SECOND

index = SoftEng_A_SECOND.index
number_of_rows_SoftEng_A_SECOND = len(index)
print('SoftEng = A, Class = SECOND, Number Of Rows = ',number_of_rows_SoftEng_A_SECOND,)
```

```python
# We repeat the process, for the instances in which SoftEng took the value 'B'. 

# Repetimos el proceso de creación de variables para SoftEng = 'B'.


print( 'FOR SPLITTING IN SoftEng = B: ')   
print('Number of Instances: ', number_instances_SoftEng_B,)
print('Number of Classes: 2')
print('Classes: FIRST, SECOND')
```

```python
# Instances in which SoftEng = B & CLASS = FIRST.

# Instancias donde SoftEng = B & CLASS = FIRST.

SoftEng_B_FIRST = pd.DataFrame(degrees[(degrees["SoftEng"] == 'B') & (degrees["Class"] == 'FIRST')])
SoftEng_B_FIRST = SoftEng_B_FIRST.reset_index(drop = True)
SoftEng_B_FIRST

index = SoftEng_B_FIRST.index
number_of_rows_SoftEng_B_FIRST = len(index)
print('SoftEng = B, Class = FIRST, Number Of Rows = ',number_of_rows_SoftEng_B_FIRST,)
```

```python
# Instances in which SoftEng = B & CLASS = SECOND.

# Instancias donde SoftEng = B & CLASS = SECOND.

SoftEng_B_SECOND = pd.DataFrame(degrees[(degrees["SoftEng"] == 'B') & (degrees["Class"] == 'SECOND')])
SoftEng_B_SECOND = SoftEng_B_SECOND.reset_index(drop = True)
SoftEng_B_SECOND

index = SoftEng_B_SECOND.index
number_of_rows_SoftEng_B_SECOND = len(index)
print('SoftEng = B, Class = SECOND, Number Of Rows = ',number_of_rows_SoftEng_B_SECOND,)
```

```python
# All variables are set. 
# We print the sum of all instances in the 'Degrees' set. 

# Todas las variables pertinentes están cargadas. 
# Imprimimos en pantalla la suma de todos los sucesos de la tabla 'Degrees'.


number_total_instances = number_instances_SoftEng_A + number_instances_SoftEng_B
print('Total Instances for the Degrees set =',number_total_instances,)
print('The starting entropy for SoftEng is ',SoftEng_Estart, 'And it will be asigned to the SoftEng_Estart variable')


print('--------------------------------------------------------')

# We print the summary of the data collected for SoftEng = A. 
# This includes total instances and per class instances, plus the entropy. 

# Imprimimos el resúmen de las variables recolectadas para SoftEng = A. 
# Esto incluye los sucesos, los sucesos por clase y la entropía. 
print('SoftEng = A')
print('Number of Instances = ',number_instances_SoftEng_A,)
print('Number of Classes = 2')
print('Class FIRST instances = ',number_of_rows_SoftEng_A_FIRST,)
print('Class SECOND instances = ',number_of_rows_SoftEng_A_SECOND,)
entropy_instances_SoftEng_A = -(number_of_rows_A_FIRST/number_instances_SoftEng_A)*math.log2(number_of_rows_A_FIRST/number_instances_SoftEng_A)-(number_of_rows_SoftEng_A_SECOND/number_instances_SoftEng_A)*math.log2(number_of_rows_SoftEng_A_SECOND/number_instances_SoftEng_A)
print('The entropy for SoftEng taking the value A is ',entropy_instances_SoftEng_A,)

print('--------------------------------------------------------')

# And the summary of SoftEng = B. 

# Y el resumen para SoftEng = B.

print('SoftEng = B')
print('Number of Instances = ',number_instances_SoftEng_B,)
print('Number of Classes = 2')
print('Class FIRST instances = ',number_of_rows_SoftEng_B_FIRST,)
print('Class SECOND instances = ',number_of_rows_SoftEng_B_SECOND,)
entropy_instances_SoftEng_B = 0
print('The entropy for SoftEng taking the value B is ',entropy_instances_SoftEng_B,)


print('--------------------------------------------------------')
# We print the evaluation of Entropies.
# For SoftEng = A and SoftEng = B by the proportion of original instances.
# The sum of the evaluated entropies will result in the variable 'SoftEng_Enew'.
# This variable contains the new entropy for the set, after splitting in SoftEng.

# Imprimimos las entropías evaluadas para SoftEng = A & SoftEng = B. 
# La suma de las entropías evaluadas es asignada a la variable 'SoftEng_Enew'
# Esta suma nos da la nueva entropía de 'Degrees', luego del split por SoftEng.

entropy_instances_SoftEng_A_EVALUATED = (number_instances_SoftEng_A/number_total_instances)*entropy_instances_SoftEng_A
print('The evaluated entropy for SoftEng taking the value A is ',entropy_instances_SoftEng_A_EVALUATED,)
entropy_instances_SoftEng_B_EVALUATED = 0
print('The evaluated entropy for SoftEng taking the value B is ',entropy_instances_SoftEng_B_EVALUATED,)
print('Both values will be added up into the variable SoftEng_Enew')
SoftEng_Enew = entropy_instances_SoftEng_A_EVALUATED + entropy_instances_SoftEng_B_EVALUATED
print('--------------------------------------------------------')


# We print SoftEng_Enew.

# Imprimimos SoftEng_Enew.

print('SoftEng_Enew = ',SoftEng_Enew,)

print('--------------------------------------------------------')

# InfoGain is the difference between SoftEng_Estart and SoftEng_Enew

# InfoGain es la diferencia entre SoftEng_Estart y SoftEng_Enew

InfoGain_SoftEng = Degrees_Estart - SoftEng_Enew

# We print InfoGain, and thus the analysis is completed. 

# Imprimimos el InfoGain, y el análisis está completado. 

print('The InfoGain for the split in SoftEng is: ',InfoGain,)
```
